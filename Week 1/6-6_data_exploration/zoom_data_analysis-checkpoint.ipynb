```python
%matplotlib inline

# Number of timesteps (rows) and labels (columns)
timesteps = emg_data.index
labels = emg_data.columns

# Setting up the figure and axis
fig, ax = plt.subplots()
bar_container = ax.bar(labels, emg_data.iloc[0])
ax.set_ylim(0, emg_data.values.max() + 1)  # Set y-axis limit for better visualization

def update(frame):
    for i, b in enumerate(bar_container):
        b.set_height(emg_data.iloc[frame, i])
    ax.set_title(f'Timestep: {frame}')
    return bar_container

# Create animation
ani = animation.FuncAnimation(fig, update, frames=range(len(timesteps)), repeat=True)

plt.show()
```
